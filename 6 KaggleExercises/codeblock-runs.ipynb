```python
# primes upto 2000000 (2 million)
import math as m
def isprime(n):
    flag = 1
    for i in range(2,m.ceil(pow(n,0.5))+1):
        if n%i==0:
            flag=0
            break
    if flag==0:
        return 0==1
    else:
        return 1==1

num = 2000000
primes = []
relsum = 0
for j in range(2,num+1):
    if isprime(j):
        relsum+=j
        primes+=[j]

print(primes[-1])
```

```python
# project euler problem 34 - sum of digits factorial
factlist = [1,1]
for i in range(2,10):
    factlist.append(factlist[-1]*i)
maxnum = factlist[-1]

ansnums = []
for j in range(3,maxnum):
    jstr = str(j)
    tmsum = 0
    for d in jstr:
        relind = int(d)
        tmsum+=factlist[relind]
    if tmsum==j:
        ansnums+=[j]

print(sum(ansnums))
```
